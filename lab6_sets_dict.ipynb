```python
low={0,1,2,3,4}
print(low) #set
odds=[1,3,5,7,9,5]
odds=set(odds)
print(odds)

odds.remove(5)
odds.add(11)

print(odds)
```

```python
#set operation
low = {0,1,2,3,4,2}
odds={1,3,5,7,9,5}
ten=set(range(10))
low.intersection(odds) #교집합
low.union(odds) #합집합
low.difference(odds) #차집합
low.issubset(ten)

print(f'{low & odds}') #intersection
print(f'{low | odds}') #union
print(f'{low - odds}') #difference

print(f'{low<=ten}') #subset
print(f'{low<ten}') #proper subset

low.clear() #공집합
```

```python
myset = {1,2,'hello', (1,2,3)} #sets are iterable object
for x in myset: #hashable한거만 집어넣을 수 있음. set은 
    print(x)
```

```python
#hashable objects
x=(1,2,3)
y=(1,2,3)

print(id(x), id(y))
print(x is y)
print(x == y)
```

```python
print(x.__hash__()) #객체 x를 “고유하게 식별하기 위한 정수값(hash 값)”을 반환하는 메서드
print(y.__hash__()) #둘이 왜 같을까? 같은 객체(값이 같음)는 항상 같은 hash
```
