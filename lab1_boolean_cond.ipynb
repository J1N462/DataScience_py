```python
height=float(input('enter height : '))/100
weight=float(input("enter weight : "))
BMI=weight/height ** 2

print("your BMI is ", BMI)

if BMI < 18:
    print("underweight")
elif BMI <= 25:
    print("normalweight")
elif BMI <= 27:
    print("overweight")
elif BMI <= 30:
    print("moderate obese")
elif BMI <= 40:
    print("severe obese")
else:
    print("very severe obese")
```
