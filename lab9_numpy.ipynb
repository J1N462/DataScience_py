```python
#list vs ndarray
import time
a=list(range(100000))
b=list(range(100000,200000))
start = time.time()
for _ in range(1000):
    c=[ i+j for i, j in zip(a,b)]
end =time.time()
print(c[:10])
print(f"{end - start:.6f} seconds")
```

```python
na = np.array(a)
nb = np.array(b)
start = time.time()
for _ in range(1000):
    nc = na + nb
end =time.time()
print(c[:10])
print(f"{end - start:.6f} seconds")
#병렬로 처리해서 빠름
# GPU로 한번에 여러 element 연산.
# GPU는 내부에 코어가 수만개씩 들어있다. -> 한번에 수만개씩 연산.
# CPU로만 계산해도 병렬 처리가 가능하다. -> SIMD instruction. 이를 위한 레지스터가 따로 있다.
# 근데 numpy는 GPU쓰는 건 아니고 SIMD를 쓴다.
```
