```python
#indexing
a=np.arange(20)
print(a)
print(a[3])
print(a[:7]) #slicing

b=a[:5]#.copy를 해줘야함
b[0]=100 #파이썬에서 슬라이싱하게되면 copy됨 / 넘파이는 안됨
#view를 만들어줌


#fancy indexing
a=np.random.randint(0,10,size=(10,)) #리스트를 갔다가 인덱스로 넘겨줄수잇
print(a)

b=a[[0,5,7,9]]
print(b)
print(a)
```

```python
#boolean indexing
a=np.random.uniform(-1,1, size=(10,))
print(a)

b=a<0 #boolean indexing용 마스크
print(b)

print(a[a<0])

a[a<0] = 0
print(a)

#boolean, fancy indexing만 카피를함
```

```python
#operation
a=np.random.randint(1,10,(5,))
b=np.random.randint(1,10,(5,))
print(a)
print(b)
print(a+b)
print(a*b) #element wise multiplicaton
print(a/b)
print(a//b) #integer division
print(a%b) #modular
print(a**2)
```

```python
#openations: broadcasting
a=np.arrange(10)
print(a)
b=a+10
print(b)
```

```python
#operation
a= np.arange(1,11)
print(a)
print(np.sin(a)) #라디안값
print(np.cos(a))
print(np.log10(a)) #log 0 == -무한대
#아무것도 안쓰면 ln이고 밑을 써주면 log n이되
```

```python
#operations
a=np.random.randint(0,100,size=(10,))
print(a)
print(np.max(a)) #function call이라는 넘파이함수불러쓰는 형태
print(np.min(a))
print(np.mean(a))
print(np.argmax(a))
print(np.argmin(a))

#obj method style
print(a.max())
print(a.min())
```

```python
#dot product
a=np.array([1,2,3])
b=np.array([3,0,1])
print(a*b) #element wise
print(np.dot(a,b)) #dot product 내적
print(a.dot(b))
print(a@b)
```
