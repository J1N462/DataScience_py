## 2D matrices, 3D arrays

```python
a=np.random.randint(0, 20, size=(3,4))
print(a)
print(a.shape)
print(a.ndim)
```

```python
#indexing
a=np.random.randint(0,20,size=(4,5))
print(a)
print(a[2,3])

b=a[:3,:3]
print(b) 
print()
b=a[::2,::2]
print(b) 

b[0,0]=100
print(a)#numpy는 copy아니고 view

c=a[:, [0,-1]]
print(c)

c[0,0]=1000 #c는 copy를 해온거라서 원본인 a는 바뀌지 않는다
print(a)
```

```python
a = np.arange(20)
print(a)
print(a.shape)
b=a.reshape(4, -1)
print(b) #reshape해도 원본수정됨
b[1,1]=1000
print(a)
#strid 얼마만큼 가면 줄을 바꿀 것인지 (row를 바꿀것인지)
```

```python
a=np.arange(10)
print(a)
print(a.shape)
b=a[:,None]
print(b)
print(b.shape)

b=a[None,:]
print(b)
print(b.shape)

b=a[np.newaxis,:,np.newaxis]
print(b)
print(b.shape)
```

```python
#c=b.reshape((10,))
c=b.reshape(-1) #view
print(c)
print(c.shape) 
```

```python
c=b.flatten()#copy
print(c)
print(c.shape)
```
