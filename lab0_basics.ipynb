## Variable in python
a nametag that points to an object.

한 object에 이름을 여러 개 붙일 수 있다.

object에 접근하려면 이름이 있어야 함.

파이썬에서 type은 object가 가지고 있는 거지 변수가 가지고 있는 게 아니다. 왜냐 변수는 그냥 이름이기 때문.

````python
welcome_msg = 'Hello world' # str object에 welcome_msg라는 이름을 붙여준다.
````

````python
print(welcome_msg) # welcome_msg refers to the string object 'Hello world' and prints it.
````

이름표가 사라진 object에는 더 이상 접근하지 못하지만, 메모리 공간에 그대로 남아있다. 이런 object를 garbage라고 한다.

한번씩 자동으로 이름표가 없는 object를 지운다. 이를 garbage collecting이라고 한다.

````python
a = 3 # a refers to the integer object 3.
print(id(a))
print(id(a))
a = 'spam' # a now refers to the string object 'spam'. The integer object 3 is still there, but it is not referred to by any name. It will be garbage collected at some point in the future.
print(id(a)) # the object that a refers to has changed, so the id is different.
````

````python
b=a # the object that a refers to is now referred to by b as well.
print(id(a))
print(id(b)) # a and b refer to the same object, so they have the same id.
````

# Operation Overloading
연산자에 다른 의미를 부여하는 것.

````python
a = 'hello'
b = 'world'
print(a + b)
print(a + ' ' + b)
# str + str => string concatenation
print(a * 3)
# 원래 산술 연산자인 +, *에 다른 의미를 부여한 것. 피연산자가 str인 경우에는 +는 문자열 연결, *는 문자열 반복을 의미한다.
````

## Python modules
- `import` 키워드로 모듈을 불러올 수 있다.

````python
#floor(2.5) error. floor is not defined. math 모듈을 import 해야 한다.
import math
math.floor(3.5)
````

````python
dir(math) # math 모듈이 제공하는 함수와 변수의 목록을 보여준다.
````

````python
from math import ceil, floor # math 모듈에서 ceil과 floor 함수만 가져온다.
#from math import * # math 모듈에서 모든 함수와 변수를 가져온다. 권장하지 않음. namespace pollution이 발생할 수 있다. 다른 모듈의 함수 또는 내가 직접 만든 이름이 겹칠 수 있다.
print(ceil(-2.5))
print(floor(-2.5))
#필요한 함수만 import하거나, 모듈 이름을 붙여서 사용하는 것을 권장한다.
````
