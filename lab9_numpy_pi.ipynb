## Image Manipulation using Numpy and PIL

```python
#numpy 비정형 데이터
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
#read an image file
pil_image = Image.open('cat_large.jpg')
np_image = np.array(pil_image) #convert to numpy array
```

```python
print(np_image.shape) #(167, 250, 3) : 0 axis, 1 axis, RGB
print(np_image.dtype) #각각에 들어있는 데이터타입. unsigned data 8
```

```python
#display
plt.imshow(np_image[:,:,:]) #original
#plt.imshow(np_image[:,::-1,:]) #horizonal flip
#plt.imshow(np_image[::-1,:,:]) #vertical flip
plt.imshow(np_image[150:500,400:750,:]) #crop
plt.imshow(np_image[::2,::2,:]) #scaling to 1/2
```

```python
pil_image=Image.fromarray(np_image[::10,::10,:])
pil_image.save("cat_scaled.jpg")
```
