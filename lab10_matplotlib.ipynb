## Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#prepare data
X=np.linspace(0,4*np.pi, 1000)
Y_sin=np.sin(X)
Y_cos=np.cos(X)
Y_tan=np.tan(X)
Y_tan[np.abs(Y_tan) > 5]=np.nan

#np.abs(Y_tan) > 5 #boolean indexing
```

```python
#render
fig, ax =plt.subplots()
ax.plot(X,Y_sin, label='sin(x)')
ax.plot(X,Y_cos, label='cos(x)')
ax.plot(X,Y_tan, label='tan(x)')
ax.set(xlabel='x', ylabel='y')
ax.grid(axis='y')
ax.legend(loc='upper right')
```

```python
#plot sigmoid
def sigmoid(x):
    return 1/(1+np.exp(-x))

X = np.linspace(-10,10,1000)
Y = sigmoid(X)

fig,ax = plt.subplots()
ax.plot(X,Y,label = r'$sigmoid(x) = \frac{1}{1+e^{-x}}$')
ax.set(xlabel = "x", ylabel = "y")
ax.set_yticks(np.linspace(0,1,11))
ax.grid(axis = 'both')
ax.legend()
plt.show()
```

```python
#scatter plot, histogram
num_points=100000
X=np.random.randn(num_points) #normal distribution 정규분포
Y=X+0.5*np.random.randn(num_points) #증가하면 y도 같이증가

fig, ax = plt.subplots(1,2,figsize=(10,5))
ax[0].scatter(X,Y, edgecolor ='k', color='r')
ax[0].set(xlabel='X', ylabel='Y')
ax[0].set_title("Scatter plot")

ax[1].hist(X, np.arange(-3,3,0.2), edgecolor='k')
ax[1].set(xlabel='X', ylabel='Frequency')
ax[1].set_title("Histogram")

fig.tight_layout()
```
