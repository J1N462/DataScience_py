```python
#read a file
reader = open('Alice.txt', 'r', encoding='UTF-8') #read
data = reader.read()
reader.close()
print(data[-100:])
```

```python
#read in chunks
reader = open('alice.txt', 'r', encoding='UTF-8')
data=reader.read(64)
while data:
    print(len(data)) 
    data = reader.read(64)
reader.close()
```

```python
#read in chunks
reader = open('alice.txt', 'r', encoding='UTF-8')
data=reader.readline()
while data:
    print(len(data)) 
    data = reader.readline()
reader.close()
```

```python
reader = open('Alice.txt', 'r', encoding='UTF-8')

num_chars, num_words, num_lines = (0,0,0)
data=reader.readline()
while data:
    num_lines += 1
    num_chars += len(data)
    num_words += len(data.split())
    data=reader.readline()


print(f"average: {num_chars/num_lines:.2f} chars per line")
print(f"average: {num_words/num_lines:.2f} words per line")
reader.close()
```

```python
#readlines
reader=open('Alice.txt', 'r', encoding='UTF-8')
data = reader.readlines()
print(len(data))
reader.close()
```

```python
#file object is iterable object
reader = open('Alice.txt', 'r', encoding='UTF-8')

num_chars, num_words, num_lines = (0,0,0)

for line in reader:
    num_lines += 1
    num_chars += len(line)
    num_words += len(line.split())


print(f"average: {num_chars/num_lines:.2f} chars per line")
print(f"average: {num_words/num_lines:.2f} words per line")
reader.close()
```

```python
#write to a file
writer = open("myfile.txt", 'w', encoding='UTF-8')
writer.write('hello world\n')
writer.writelines('hello\n', 'good\n', 'morning\n')
writer.close()
```

#binary files : b 붙이기

```python
#with statement
with open('alice.txt', 'w', encoding='utf-8') as f:
    f.write('hello')
```

```python
#copy a file
with open('alice.txt', 'r', encoding='utf-8') as reader:
    with open('alice_copied.txt', 'w', encoding='utf-8') as writer:
        for line in reader:
            writer.write(line)
        
```
