## Analysis wit plots

```python
#visualize the distribution of ages
ages = df['Age']
ax = ages.hist(bins=20, edgecolor = 'k')
ax.set(xlabel='age', ylabel="counts")
ax.set_title("Age Distribution")
```

```python
#plot 'income' histogram
income = df['Income']
#print(ages)

ax = income.hist(bins = 20, edgecolor = 'k')
ax.set(xlabel='income', ylabel='columns')
ax.set_title('Distribution of income')
```

```python
fig, axes = plt.subplots(1,2,figsize=(10,5))

ages.hist(ax=axes[0],bins=20, edgecolor='k',color='red')
ax.set(xlabel='age', ylabel="counts")
ax.set_title("Age Distribution")

income.hist(ax=axes[1],bins=20, edgecolor='k',color='blue')
ax.set(xlabel='income', ylabel='columns')
ax.set_title('Distribution of income')
```

## Sorting and Filtering

```python
# df.sort_values 
# by 어떤 컬럼을 갖다가 기준으로 sorting할지
#inplace
df_sorted_by_age=df.sort_values(by='Age', ascending=True)
df_sorted_by_age
```

```python
df_sorted=df.sort_values(by=['Age','Income'], ascending=[False, True])
df_sorted[['Age','Income']]
```

```python
#filtering
df["Gender"].unique()
```

```python
# Q : what percentage of females?
females=df["Gender"]=='female'
females
df[females]
print(f"females : {len(df[females]) / len(df) * 100 : .01f} %")
```

```python
males=df["Gender"]=='male'
print(f"males : {len(df[males]) / len(df) * 100 : .01f} %")
```

```python
#fig, axes = plt.subplots(1,2,figsize=(10,4))
#df[females]['Age'].hist(ax=axes[0], bins=20, color = 'red', edgecolor='k')

fig, axes = plt.subplots()
ax = df[males]['Age'].hist(bins = 20, ax = axes, color = 'blue', alpha = 0.6, edgecolor = 'k',label = 'male ages')
ax = df[females]['Age'].hist(bins = 20, ax = axes, color = 'r', alpha = 0.6, edgecolor = 'k', label = 'female ages')

ax.set(xlabel='age', ylabel='columns')
ax.set_title('Male/Female Age Distribution Difference')
axes.legend()
```
