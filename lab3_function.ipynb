```python
#object가 타입이 있고 변수는 타입이 없다
#input type를 강제할 수 없다. type를 명시하고 싶다면 주석을 달아라

#defining a function
def double(x):
    '''
    calculate a double of 'x'
    Args:
        x: int input
    Returns:
        twice of x
    '''
    
    return 2*x
```

```python
print(double(30))
print(double('hello'))
print(double([1,2,3,'hello']))

#함수 정의할 때 쓰는 변수 이름 : parameter (매개변수)
#함수 호출할때 parameter에 실제로 할당하는 값 : argument (인자)
```

```python
def print_square(size : int):
    for _ in range(size):
        for _ in range(size):
            print("#", end="")
        print("") 
```

```python
print_square(4)
```

```python
#function 또한 object이다, 데이터가 아닌 코드가 들어있는 오브젝트
ps=print_square
ps(3)
#함수 스택프레임 : 변수나 함수나 이런것들이 공간을 갖는데 그게 스택처럼 호출하면 push(), return하면 pop()
```

```python
'''
def factorial(x):
    fact=1
    while x > 1:
        fact = fact * x
        x=x-1
    return fact
'''
def factorial(x):
    if x == 1:
        return 1
    else:
        return x*factorial(x-1)
        
```

```python
factorial(3)
```

```python
#lab: sum from 1 to n
def sum_to_num(num):
    if num == 1:
        return 1
    else:
        return num + sum_to_num(num - 1)
```

```python
sum_to_num(10)
```

```python
#default value for parameters
def my_func(name, age=20, country='Norway'):
    print(name, "is from", country, age)
    
my_func("Inho")
my_func('Inho', country='Korea', age='40')
```

```python
from math import pi

def perimeter(radius):
    return 2*pi*radius

def area(radius):
    return pi*radius**2

my_circle_area = area #이름이 두개가됨
my_funcs = [perimeter, area, my_circle_area]

for f in my_funcs:
    print(f.__name__, f(2.87))
    #printf(f'{function.__name__}:{function(2.87):.2f}')
    #__name__은 함수(또는 객체)의 이름을 문자열로 알려주는 속성
    #
```
