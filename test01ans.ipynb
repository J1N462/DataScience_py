```python
grades=[
    ('bob', 'math', 80),    
    ('alice', 'math', 60),    
    ('bob', 'english', 90),    
    ('alice', 'english', 80),    
    ('bob', 'history', 70),    
    ('alice', 'history', 90)
]
```

```python
grades.sort()
for x in grades:
    print(x)
grades.sort(key=lambda x:x[2], reverse=True)

print()
for x in grades:
    print(x)
```

```python
def subject_averages(grades):
    sum={}
    count={}
    avg={}
    
    for _,s,g in grades:
        sum[s]=sum.get(s,0) +g
        count[s]=count.get(s,0)+1
        
    for s in sum.keys():
        avg[s]=sum[s]/count[s]
    return avg

print(subject_averages(grades))
```

```python
grades.sort(key=lambda x:x[2], reverse=True)
grades.sort(key=lambda x:x[0])

for x in grades:
    print(x)
```
