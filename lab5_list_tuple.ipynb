```python
fruits =  ["apple","kiwi", "banana", "mango", "tomamo"]
fruits.sort(reverse=True, key=len)
print(fruits)
```

```python
def last_char(s):
    return s[-1] #lambda function이랑 같은역할

fruits =  ["apple","kiwi", "banana", "mango", "tomamo"]
fruits.sort(key=last_char)
print(fruits)
fruits.sort(key=lambda x: x[-1])
print(fruits)
```

```python
def double(mylist):
    for index, value in enumerate(mylist):
        mylist[index] = 2*value
        
things = [2,5,'spam', 9.5]
double(things)
print(things)
```

```python
def double(seq):
    seq=seq*2
things=10
double(things)
print(things)
# things → 정수 (immutable)
# 함수 안에서 seq = seq * 2는 새로운 객체(20)를 만들고 seq가 그걸 가리킴
#원래 things는 그대로 (10 유지)

#함수 내부에서 재할당만 일어나고 원본은 안 바뀜
```

## List comprehension

```python
my_list=[i**2 for i in range(10)]
print(my_list)
```

```python
li=[('a', 1), ('c', 3), ('d', 2)]

[[n for k, n in li]]
```

```python
#split the string and make a list of int string
num_string='1 2 3 4 5 6'
li=[int(i) for i in num_string.split()]
print(li)
```

```python
#exchange two num in each tuple
num_pairs=[(1,2), (3,4), (5,6)]
li=[(j,i) for i , j in num_pairs]
print(li)
```

```python
#교수님 1번 풀이
num_string='1,2,3,4,5,6'
my_list=num_string.split(',')
print(my_list)
my_num_list=[int(c) for c in my_list]
print(my_num_list)
```

```python
li=[3,6,2, 7,1, 9]
lili=[i for i in li if i%2==0]
print(lili)

lilili=[n**2 for n in li if n%2==0]
print(lilili)

lililili=[n if n%2==0 else 0 for n in li]
print(lililili)
```

```python
nums=[0, 1, -3, -5, 3, 6, 2]
num_positive=[i if i>0 else -i for i in nums ]
print(num_positive)
```
