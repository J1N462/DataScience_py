```python
fruits =  ["apple","kiwi", "banana", "mango", "tomamo"]
fruits.sort(reverse=True, key=len)
print(fruits)
```

```python
def last_char(s):
    return s[-1] #lambda function이랑 같은역할

fruits =  ["apple","kiwi", "banana", "mango", "tomamo"]
fruits.sort(key=last_char)
print(fruits)
fruits.sort(key=lambda x: x[-1])
print(fruits)
```

```python
alpha=[]
for i in range(ord('a'), ord('z')+1):
    alpha.append(chr(i))
print(alpha)
alpha.reverse()
print(alpha)
```
