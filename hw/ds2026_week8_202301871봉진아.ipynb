```python
import numpy as np
def tri_i(n=5):
    X=np.arange(0,n*n).reshape(n,n)
    a=np.arange(0,n)
    b=np.arange(0,a)
    X=True
    X[b:a]=False
    return X

#tri_i(4)

def my(x, size):
    return np.mean(x[::size, ::size])

x=np.arange(1,17).reshape(4,4)
print(my(x,2))

np.full((2,3,4), 5, dtype=np.float32)
```
