```python
#6
def Roma(num):
    Roman=['I', 'V', 'X']
    if 1 <= num <= 3:
        return Roman[0] * num
    elif num == 4:
        return Roman[0] + Roman[1]
    elif num == 5:
        return Roman[1]
    elif 6 <= num <= 8:
        return Roman[1] + Roman[0] * (num - 5)
    elif num == 9:
        return Roman[0] + Roman[2]
    elif num == 10:
        return Roman[2]
    
print(Roma(7))
```

```python
#7
R_num=[]
for i in range(1,11):
    R_num.append(Roma(i))
print(R_num)
```

```python
#8
for idx, val in enumerate(R_num, start=1):
    print(idx, ":", val)
```
