```python
#4
info = {
    "high_school" : {
        '1' : ['sunwoo', 'lee'],
        '2' : 'kim',
        '3' : 'park'
    },
    "middle_school" : {
        "2" : ['andrew', 'yoshua']
    }
}
school_data = {}

class Classroom:
    def __init__(self, school, grade, name):
        self.school=school
        self.grade = str(grade)

        if isinstance(name, list):
            self.name = name
        else:
            self.name = [name]
        
    def add_student(self, grade, name):
        new_grade=str(grade)
        if self.grade != new_grade:
            return "Invalid input"
        elif isinstance(name, list):
            self.name.extend(name)
        else:
            self.name.append(name)
        
    def __len__(self):
        return len(self.name)   
         
    def __str__(self):  
        return f"{self.grade}_Grade : {self.name}"

for school in info:
    school_data[school] = {}
    for grade in info[school]:
        school_data[school][grade] = Classroom(
            school,
            grade,
            info[school][grade]
        )
    
for i in school_data:    
    print(i)    
    for j in school_data[i]:        
        print(school_data[i][j])
```

```python
#5
high_school_num = [{f"{grade}grade": len(school_data["high_school"][grade])} 
                   for grade in school_data["high_school"]]

middle_school_num = [{f"{grade}grade": len(school_data["middle_school"][grade])} 
                     for grade in school_data["middle_school"]]

print(high_school_num)
print(middle_school_num)
```

```python
#6
h_sum=sum(list(d.values())[0] for d in high_school_num)
m_sum=sum(list(d.values())[0] for d in middle_school_num)

print(f'high_school : {h_sum}')
print(f'middle_school : {m_sum}')
```
