```python
#16
mail = input("enter string(8-alphabet) : ")
count=0

for char in mail:
    if char=="M" or char=="S":
        count+=1

if(count>=5):
    print("Secret mail!")
else:
    print("평범한 문자")
```

```python
#20-1
from math import cos, pi
import random
co=cos(pi)
print("cos(pi) is", str(co))

#20-2
ran=random.randint(-10,10)
if ran>co:
    print("random_value is "+ str(ran) +" and bigger than "+ str(co))
elif ran==co:
    print("random_value is "+ str(ran) +" and same with "+ str(co))
else:
    print("random_value is "+ str(ran) +" and smaller than "+ str(co))
```
