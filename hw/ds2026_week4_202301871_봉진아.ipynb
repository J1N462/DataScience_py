```python
#3
import time

def check_UTC(unit):
    now = time.gmtime()
    
    if unit == "연도":
        print(now.tm_year, "\nyear")
    elif unit == "월":
        print(now.tm_mon, "\nmonth")
    elif unit == "일":
        print(now.tm_mday, "\nday")
    elif unit == "시간":
        print(now.tm_hour, "\nhour")
    elif unit == "분":
        print(now.tm_min, "\nminute")
    elif unit == "초":
        print(now.tm_sec, "\nsecond")
    else:
        print("오류")
        
time_part=input()
check_UTC(time_part)
```

```python
#4
def recursion_power(n):
    total = 0
    for i in range(1, n+1):
        total += i**2
    return total

print(recursion_power(4))
```

```python
#5
def loop_power(n):
    total = 0
    for i in range(1, n + 1):
        total += i ** 2
    return total

print(loop_power(4))
```

```python
#6
def check_func_time(func, num):
    """
    Execute the given function and measure its execution time in nanoseconds.

    Args :
        func : 
            The function to measure.
        num : 
            The argument to pass to func.

    Returns
        The execution time of the function in nanoseconds.
    """
    start_point = time.perf_counter_ns()
    result = func(num)
    print(result)
    end_point = time.perf_counter_ns()
    return end_point - start_point

func_work_time = check_func_time(loop_power, 5)
print(func_work_time)
```

```python
#7
lps = loop_power
print(lps(5))
# 함수는 파이썬에서 "오브젝트"이기 때문에 변수에 저장할 수 있다. 
# lps변수는 기존 loop_power함수를 가리키게 된다.

print(lps.__name__)
# __name__은 함수 오브젝트가 원래 가지고 있는 이름을 보여주는데,
# lps는 단순히 loop_power를 참조만 하고 있어서 출력결과 원래 함수 이름인 "loop_power"가 나온다.
```

```python
#8-1
import time

power_sum = [loop_power, recursion_power]

#8-2
result_time=0.0
iter=5
for func in power_sum:
    total_time = 0
    
    for _ in range(iter):
        start = time.perf_counter_ns()
        print(func(5))
        end = time.perf_counter_ns()
        total_time += (end - start)
    
    print(func.__name__, "평균 시간:", total_time / iter)

```

```python
#10
def unfrog(x):
    if type(x) == int:
        print(chr(x))
    else:
        print(ord(x))
        
unfrog(77)
```

```python
#11
def ununfrog(x):
    if type(x) == str:
        uni = ord(x)
        print(bin(uni))
        print(oct(uni))
        print(hex(uni))
    else:
        print(chr(x).encode("utf-16"))
        
ununfrog('B')
```
