### plot sigmoid and relu function in a single figure

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

X = np.linspace(-10,10,100)
Y = sigmoid(X)


def relu(x):
    return np.maximum(0,x)

A = np.linspace(-10,10,100)
B = relu(X)
```

```python
fig,ax=plt.subplots(1,2,figsize=(10,4))
ax[0].plot(X,Y,label = r'sigmoid(x)')
ax[0].set(xlabel = "x", ylabel = "y")
ax[0].set_yticks(np.linspace(0,1,11))
ax[0].grid(axis = 'both')
ax[0].legend()

ax[1].plot(A,B,label=r'relu(x)')
ax[1].set(xlabel='x', ylabel='y')
ax[1].set_yticks(np.linspace(0,10,6))
ax[1].grid(axis = 'both')
ax[1].legend()

plt.show()
```
