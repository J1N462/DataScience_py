```python
#11
reader=open('Atoz.txt', 'r', encoding='utf-8')
alpha={}
txt=reader.read().lower()
for i in txt:
    alpha[i]=0

print(alpha)
```

```python
#12
with open('the-blue-hotel.txt', 'r', encoding='utf-8')as reader:
        txt=reader.read().lower()
        for ch in txt:
            if ch in alpha:
                alpha[ch]+=1
print(alpha)
```

```python
#13
max_alpha=max(alpha, key=alpha.get)
min_alpha=min(alpha, key=alpha.get)

print(max_alpha,alpha[max_alpha])
print(min_alpha,alpha[min_alpha])
```

```python
#13-1
max3=sorted(alpha.items(),key=lambda x:x[1], reverse=True)[:3]
min3=sorted(alpha.items(),key=lambda x:x[1])[:3]

for k,v in max3:
    print(k,v)
for k,v in min3:
    print(k,v)
```
