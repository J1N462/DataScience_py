```python
#1
class shapes:
    def __init__(self, x, y, name):
        self.x = x
        self.y = y
        self.name = name
        
        
    def get_area(self):
        if self.name == "triangle":
            return (self.x * self.y) / 2
        elif self.name == "rectangle":
            return self.x * self.y
        else:
            return 0
        
    def __str__(self):
        return f"{self.name} area: {self.get_area()}"
        
    def _check(self, other):
        return isinstance(other, shapes)

    def __add__(self, other):
        if not self._check(other):
            return "Invalid operations"
        return self.get_area() + other.get_area()

    def __sub__(self, other):
        if not self._check(other):
            return "Invalid operations"
        return abs(self.get_area() - other.get_area())

    def __mul__(self, other):
        if not self._check(other):
            return "Invalid operations"
        return self.get_area() * other.get_area()

    def __truediv__(self, other):
        if not self._check(other):
            return "Invalid operations"
        if other.get_area() == 0:
            return "Invalid operations"
        return self.get_area() / other.get_area()

    def __mod__(self, other):
        if not self._check(other):
            return "Invalid operations"
        return self.get_area() % other.get_area()

    


t = shapes(30, 40, "triangle")
print(t)
    
```

```python
#4
class shapes:
    def __init__(self, x, y, name):
        self.x = x
        self.y = y
        self.name = name

    def get_area(self):
        if self.name == "triangle":
            return (self.x * self.y) / 2
        elif self.name == "rectangle":
            return self.x * self.y
        return 0

    def __str__(self):
        return f"{self.name} (x={self.x}, y={self.y})"

    def __add__(self, other):
        if isinstance(other, shapes):
            return self.get_area() + other.get_area()
        return shapes(self.x + other, self.y + other, self.name)

    def __radd__(self, other):
        return self.__add__(other)

    def __sub__(self, other):
        if isinstance(other, shapes):
            return abs(self.get_area() - other.get_area())
        return shapes(abs(self.x - other), abs(self.y - other), self.name)

    def __rsub__(self, other):
        return shapes(abs(other - self.x), abs(other - self.y), self.name)

    def __mul__(self, other):
        if isinstance(other, shapes):
            return self.get_area() * other.get_area()
        return shapes(self.x * other, self.y * other, self.name)

    def __rmul__(self, other):
        return self.__mul__(other)

    def __truediv__(self, other):
        if isinstance(other, shapes):
            return self.get_area() / other.get_area()
        return shapes(self.x / other, self.y / other, self.name)

    def __rtruediv__(self, other):
        return shapes(other / self.x, other / self.y, self.name)

    def __mod__(self, other):
        if isinstance(other, shapes):
            return self.get_area() % other.get_area()
        return shapes(self.x % other, self.y % other, self.name)

    def __rmod__(self, other):
        return shapes(other % self.x, other % self.y, self.name)



# 테스트
t = shapes(30, 40, "triangle")
r = shapes(20, 70, "rectangle")

print(t + 10)     
print(t - 24)     
print(t * 3)
print(t / 2)
print(t % 5)
print(t + r)  
```
