```python
def overlap(ax1, ay1, ax2, ay2, bx1, by1, bx2, by2):
    # 좌하단 중 더 큰 값
    overlap_x1 = max(ax1, bx1)
    overlap_y1 = max(ay1, by1)
    
    # 우상단 중 더 작은 값
    overlap_x2 = min(ax2, bx2)
    overlap_y2 = min(ay2, by2)
    
    width = overlap_x2 - overlap_x1
    height = overlap_y2 - overlap_y1
    
    if width > 0 and height > 0:
        area = width * height
        print("overlap!")
        print(f"area : {area}")
        return area
    else:
        print("no_overlap!")
        return 0
```

```python
# Case #1
ax1, ay1, ax2, ay2 = 1, 2, 4, 5
bx1, by1, bx2, by2 = 2, 1, 5, 4

result = overlap(ax1, ay1, ax2, ay2, bx1, by1, bx2, by2)
```
